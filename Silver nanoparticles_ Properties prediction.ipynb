```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

SilverNPs = pd.read_csv("../input/SilverNPs.csv")
```

```python
SilverNPs.describe()
```

There are columns filled with zero values. 

```python
df=SilverNPs.loc[:, (SilverNPs != 0).any(axis=0)]
```

```python
df.describe()
```

Dropping columns with more than 15% zeros.

```python
drop_cols = df.columns[(df == 0).sum() > 0.15*df.shape[0]]
df.drop(columns=drop_cols, inplace = True) 
```

```python
df.info()
```

ID, Std_Dev, Max and Min can be dropped out. R_avg, R_min, R_max, R_diff and R_std can be dropped out, because the radius of the nanoparticle depends on number of atoms and coordination number.

```python
df.drop(columns = ['ID','Std_Dev', 'Max', 'Min','R_avg', 'R_max', 'R_min', 'R_diff', 'R_std'], inplace = True) 
```

The correlation between the features and the target variable "Formation_E'

```python
corr_matrix = df.corr()
print(corr_matrix["Formation_E"].sort_values(ascending=False))
```

The 4 most relevant features for the analysis (not including EA, EF, EG, IP because they are target variables as well) are:
Avg_total, Avg, Avg_surf and N_Atoms. La energia debe depender del numero de atomos y como se acomodan para formar la particula.

```python
feat_cols=['Avg_total', 'Avg', 'Avg_surf', 'N_Atoms']
```

Plots of Formation energy as a function of high correlation variables ORIGINAL DATASET

```python
plt.figure(figsize=(15, 8))
plt.subplot(221)
plt.plot(df['Avg_total'], df['Formation_E'], '.', c='g')
plt.xlabel('Average Ag coordination number')
plt.ylabel('Formation energy (eV)')

plt.subplot(222)
plt.plot(df['Avg'], df['Formation_E'], '.', c='m')
plt.xlabel('Average bond length (A)')
plt.ylabel('Formation energy (eV)')

plt.subplot(223)
plt.plot(df['Avg_surf'], df['Formation_E'], '.')
plt.xlabel('Average surface Ag coordination number')
plt.ylabel('Formation energy (eV)')

plt.subplot(224)
plt.plot(df['N_Atoms'], df['Formation_E'], '.', c='b')
plt.xlabel('Number of atoms')
plt.ylabel('Energy of the Fermi level')


plt.show()
```

Dataframe with the relevant features and the target variable Formation Energy.

```python
X=df[feat_cols]
y=df['Formation_E']
#df_1=df[['Formation_E']+feat_cols]
```

```python
from sklearn.model_selection import train_test_split
X_1, X_val, y_1, y_val = train_test_split(X,y, test_size=0.65, random_state=0)
```

```python
X_1.shape, y_1.shape, X_val.shape, y_val.shape
```

```python
fig=plt.figure(figsize=(15, 8))

plt.subplot(221)
plt.plot(X_1['Avg_total'], y_1, '.', c='g')
plt.xlabel('Average Ag coordination number')
plt.ylabel('Formation energy (eV)')


plt.subplot(222)
plt.plot(X_1['Avg'], y_1, '.', c='m')
plt.xlabel('Average bond length (A)')
plt.ylabel('Formation energy (eV)')

plt.subplot(223)
plt.plot(X_1['Avg_surf'], y_1, '.')
plt.xlabel('Average surface Ag coordination number')
plt.ylabel('Formation energy (eV)')

plt.subplot(224)
plt.plot(X_1['N_Atoms'], y_1, '.', c='b')
plt.xlabel('Number of atoms')
plt.ylabel('Energy of the Fermi level')
plt.show()
```

The distribution of points is uniform.

**Cross-Validation and model selection**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score

def cv_scores(model, x,y):
    scores=cross_val_score(model, x, y, cv=5)
    return "Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2)
```

Dado el numero de samples en el dataset, usaremos esos parametros para no overfittear. Differentes n_estimators fueron usados para el RFR, pero no valieron madres.

```python
print(cv_scores(RandomForestRegressor(n_estimators=10), X_1, y_1))
print(cv_scores(KNeighborsRegressor(n_neighbors=5), X_1, y_1))
```

The model of choice is Random Forest Regressor. It will be trained with the whole training dataset X_1 and y_1, and then tested agains the unseen dataset X_val and y_val.

```python
# Training the model

model_RF=RandomForestRegressor(n_estimators=10)
model_RF.fit(X_1, y_1)
```

```python
model_RF.score(X_1, y_1)
```

```python
from sklearn.metrics import mean_absolute_error
y_predict=model_RF.predict(X_val)
mae = mean_absolute_error(y_predict, y_val)
print(mae)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(221)
plt.plot(X_val['Avg_total'], y_val, '+', c='b')
plt.plot(X_val['Avg_total'], y_predict,'.', c='r')
plt.xlabel('Average Ag coordination number')
plt.ylabel('Formation energy (eV)')


plt.subplot(222)
plt.plot(X_val['Avg'], y_val, '+', c='b')
plt.plot(X_val['Avg'], y_predict, '.', c='r')
plt.xlabel('Average bond length (A)')
plt.ylabel('Formation energy (eV)')

plt.subplot(223)
plt.plot(X_val['Avg_surf'], y_val, '+', c='b')
plt.plot(X_val['Avg_surf'], y_predict, '.', c='r')
plt.xlabel('Average surface Ag coordination number')
plt.ylabel('Formation energy (eV)')

plt.subplot(224)
plt.plot(X_val['N_Atoms'], y_val, '+', c='b')
plt.plot(X_val['N_Atoms'], y_predict,'.', c='r')
plt.xlabel('Number of atoms')
plt.ylabel('Energy of the Fermi level')


plt.show()
```
